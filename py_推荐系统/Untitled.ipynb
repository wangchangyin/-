```python
#采用python字典来表示每位用户的音乐和次数
file = open("data.csv",'r', encoding='UTF-8')#记得读取文件时加‘r’，以只读方式打开文件。 encoding='UTF-8'
##读取data.csv中每行中除了名字的数据
data = {}##存放每位用户评论的电影和评分
for line in file.readlines()[1:92835]:
    line = line.strip().split(',')
    line[1]=(int(line[1]))
    #如果字典中没有某位用户，则使用用户ID来创建这位用户
    if not line[0] in data.keys():
        data[line[0]] = {line[3]:line[1]}
    #否则直接添加以该用户ID为key字典中
    else:
        data[line[0]][line[3]] = line[1]
 
#print(data)


#计算任何两位用户之间的相似度，由于每位用户音乐不完全一样，所以兽先要找到两位用户共同评论过的电影
#最后算出两者之间的相似度
from math import *
def Euclidean(user1,user2,user1_num,user2_num):
    #取出两位用户音乐次数
    user1_data=data[user1]
    user2_data=data[user2]
    
    
    #计算相似度 #找到两位用户都评论过的音乐
    similar1=0
    similar2=0
    similar3=0
    for key in user1_data.keys():
        if key in user2_data.keys():
            user1_data[key]=int(user1_data[key])
            similar1+=(user1_data[key]-user1_num)*(user2_data[key]-user2_num)
            similar2+=pow((user1_data[key]-user1_num),2)
            similar3+=pow((user2_data[key]-user2_num),2)
    similar2=float(sqrt(similar2))
    similar3=float(sqrt(similar3))
    if (similar2*similar3)==0:
        return 0
    
    return float(similar1/(similar2*similar3))

#     distance = 0
#     #找到两位用户都评论过的音乐
#     for key in user1_data.keys():
#         if key in user2_data.keys():
#             #注意，distance越大表示两者越相似
#             #print(user1_data[key])
#             distance += pow(float(user1_data[key])-float(user2_data[key]),2)
#     return 1/(1+sqrt(distance))#这里返回值越小，相似度越大
 
#计算某个用户与其他用户的相似度
def top10_simliar(userID):
    res = []
    for userid in data.keys():
        #排除与自己计算相似度
        if not userid == userID:
            #计算平均值
            user1_data=data[userID]
            user2_data=data[userid]
            i=0
            j=0
            #算出用户的平均值
            user1_num=0
            user2_num=0
            for key in user1_data.keys():
                user1_num += int(user1_data[key])
                i=i+1
            for key in user2_data.keys():
                user2_num += int(user2_data[key])
                j=j+1
            user1_num/=i
            user2_num/=j
            
            simliar = Euclidean(userID,userid,user1_num,user2_num)
            res.append((userid,simliar,user1_num,user2_num))
    res.sort(key=lambda val:val[1],reverse=True)
    return res[:4]
 
# RES = top10_simliar('2')
# print(RES)
```
